```python
#boilerplate
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
%config InlineBackend.figure_format='svg'
import scipy.integrate as integrate
```

### **Problem B1**

**a)**

The probability that the sample's strength is 6250$\frac{Kg}{cm^2}$ or lower means a Z-score of 2.5 since a standard deviation is 100$\frac{Kg}{cm^2}$ and the mean is 6000$\frac{Kg}{cm^2}$. According to the Z tables, this value of less than 2.5$\sigma$ would include 99.38% of samples. We can confirm this by integrating using the built-in scipy package:

```python
sigma=100
mu=6000
fx=lambda x:1/np.sqrt(2*np.pi*sigma)*np.exp(-(x-mu)**2/(2*sigma**2))
result=integrate.quad(fx,0,6250)
print(0.1*result[0])
```

Both methods give an identical answer of 99.38%. This means that the vasy majority of the time, a sample will fall below the cutoff of 6250$\frac{Kg}{cm^2}$.

**b)**

We now wish to find what proportion of samples fall between 5800$\frac{Kg}{cm^2}$ and 5900$\frac{Kg}{cm^2}$. We can repeat the same process as before, just changing the bounds of integration to suit this problem.

```python
sigma=100
mu=6000
fx=lambda x:1/np.sqrt(2*np.pi*sigma)*np.exp(-(x-mu)**2/(2*sigma**2))
result=integrate.quad(fx,5800,5900)
print(0.1*result[0])
```

We find that 13.59% of samples fall within this range.

**c)**

If we want to find the strength exceeded by 95% of samples, we can look at the Z tables to find that this corresponds to $\approx$ +1.65$\sigma$. Plugging in our original $\mu$ and $\sigma$ gives:

$$
6000\frac{Kg}{cm^2}+1.65*100\frac{Kg}{cm^2}=\boxed{6165\frac{Kg}{cm^2}}
$$

We can also confirm this mathematically by integrating the normal distribution:

$$
\int_{s}^{\infty}\frac{1}{\sqrt{2\pi*100}}e^{\frac{-(x-6000)^2}{2*100^2}}dx=0.95
$$

Solving gives an answer of 6131$\frac{Kg}{cm^2}$.

### **Problem B2**

```python
rolls=[]
i=0
while i<1001:
    result1=np.random.randint(6)+1
    result2=np.random.randint(6)+1 #generate 2 random integers from 1 to 6
    #print(result) #show each dice roll
    sumroll=result1+result2 #sum each dice roll
    rolls.append(sumroll)
    i=i+1
plt.hist(rolls,bins=11)
plt.title('Frequency of Sum of Two Dice Rolls')
plt.ylabel('Number of Occurrences')
plt.xlabel('Sum of Two Dice Rolls')
print()
```
